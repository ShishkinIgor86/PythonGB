Задача №67. Решение в группах
1. Создать новый столбец в таблице с
пингвинами, который будет отвечать за
показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)

```python
penguins.loc[penguins['bill_length_mm']<35, 'length_class'] = 'Маленький'
penguins.loc[(penguins['bill_length_mm']>=35) & (penguins['bill_length_mm']<=42), 'length_class'] = 'Средний'
penguins.loc[penguins['bill_length_mm']>42, 'length_class'] = 'Большой'
penguins.head(45)
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data_1=data.copy()
data_2=data.copy()
data.head(20)
```

Первый вариант:

```python
def h_r(x):
  if x=='robot' or x==1:
    return 0
  else:
    return 1
```

```python
data_1['whoAmI_human']=data_1['whoAmI'].apply(h_r)
data_1['whoAmI_robot']=data_1['whoAmI_human'].apply(h_r)
data_1.head(20)
```

Второй вариант:

```python
l_d=data_2.values.tolist()
h_d=data_2.whoAmI.unique()
res=[None for i in range(len(l_d))]
res1=[None for i in range(len(l_d))]
for i in range(len(l_d)):
  if h_d[0] in l_d[i]:
    res[i]=1
    res1[i]=0
  else:
    res[i]=0
    res1[i]=1

res_data=pd.DataFrame({'whoAmI':l_d,
                       'whoAmI'+'_'+str(h_d[0]):res,
                       'whoAmI'+'_'+str(h_d[1]):res1})
res_data.head(20)
```
